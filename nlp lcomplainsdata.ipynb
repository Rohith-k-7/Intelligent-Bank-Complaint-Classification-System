```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import re

def normalize_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)   
    text = re.sub(r"@\w+", "", text)                      
    text = re.sub(r"#\w+", "", text)              
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    tokens = word_tokenize(text)              # tokenize
    # Lemmatization
    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]             
    return " ".join(lemmatized_words)

df["Text"] = df["Text"].apply(normalize_text)
df.head()
```

```python
df['Product'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df['Product']=le.fit_transform(df['Product'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['Text'],
    df['Product'],
    test_size=0.2,
    random_state=42,
    stratify=df['Product']
)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=20000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

max_len = 300
X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding='post')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dropout,Dense

model = Sequential([
    Embedding(input_dim=20000, output_dim=128,input_length=max_len),
    LSTM(256),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(len(le.classes_), activation='softmax')
])

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    X_train_pad,
    y_train,
    epochs=6,
    batch_size=32,
    validation_data=(X_test_pad, y_test)
)
```

```python
def analyze_complaint_all_classes(complaint_text):
    # Step 1: Text to sequence
    seq = tokenizer.texts_to_sequences([complaint_text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')

    # Step 2: Model prediction
    predictions = model.predict(pad)[0]   # shape: (num_classes,)

    # Step 3: Decode class names
    class_labels = le.inverse_transform(range(len(predictions)))

    # Step 4: Create result dictionary
    class_probabilities = {
        class_labels[i]: float(predictions[i])
        for i in range(len(predictions))
    }

    # Step 5: Best class
    best_class = max(class_probabilities, key=class_probabilities.get)

    return {
        "All Class Probabilities": class_probabilities,
        "Final Predicted Class": best_class
    }
```

```python
import pickle
import json
import re
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Load model
model = load_model("complaint_lstm_model.keras")

# Load tokenizer
with open("tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

# Load label encoder
with open("label_encoder.pkl", "rb") as f:
    le = pickle.load(f)

# Load params
with open("model_params.json", "r") as f:
    params = json.load(f)

max_len = params["max_len"]   # = 300 ✅
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    return text
```

```python
def analyze_complaint_all_classes(complaint_text):
    text = clean_text(complaint_text)

    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')

    predictions = model.predict(pad)[0]  # shape: (num_classes,)

    results = {}
    for label, prob in zip(le.classes_, predictions):
        results[label] = round(float(prob), 4)

    return results
```

```python
def predict_complaint(complaint_text):
    text = clean_text(complaint_text)
    seq = tokenizer.texts_to_sequences([text])
    pad = pad_sequences(seq, maxlen=max_len, padding='post')

    pred = model.predict(pad)
    class_index = pred.argmax(axis=1)[0]

    return le.inverse_transform([class_index])[0]
```

```python
analyze_complaint_all_classes(
    "UPI transaction is showing pending for more than two days"
)
```

```python
predict_complaint(
    "UPI transaction is showing pending for more than two days"
)
```
